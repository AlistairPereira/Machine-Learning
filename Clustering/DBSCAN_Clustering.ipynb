```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
x= np.array([[1,2],[2,2],[2,3],[8,7],[8,8],[25,80]])
```

```python
x
```

```python
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
db = DBSCAN(eps=3,min_samples=2)
db.fit(x)
db.labels_
```

```python
db = DBSCAN(eps=10,min_samples=3)
db.fit(x)
db.labels_
```
